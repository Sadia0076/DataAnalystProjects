```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics
import scipy.stats as stats
```

2. Load Dataset

```python
data = pd.read_csv('/content/Mall_Customers.csv')
```

3. Data Information & Cleaning

```python
data.info()
data.isnull().sum()
```

If missing values

```python
data = data.dropna()  # missing fill values
```

4. Descriptive Statistics

```python
# Mean
print("Mean Age:", data['Age'].mean())
print("Mean Annual Income:", data['Annual Income (k$)'].mean())
print("Mean Spending Score:", data['Spending Score (1-100)'].mean())

# Median
print("Median Age:", data['Age'].median())
print("Median Annual Income:", data['Annual Income (k$)'].median())
print("Median Spending Score:", data['Spending Score (1-100)'].median())

# Mode
print("Mode Age:", data['Age'].mode()[0])
print("Mode Annual Income:", data['Annual Income (k$)'].mode()[0])
print("Mode Spending Score:", data['Spending Score (1-100)'].mode()[0])
```

Measures of Dispersion:

```python
# Variance
print("Variance Age:", data['Age'].var())
print("Variance Annual Income:", data['Annual Income (k$)'].var())
print("Variance Spending Score:", data['Spending Score (1-100)'].var())

# Standard Deviation
print("Standard Deviation Age:", data['Age'].std())
print("Standard Deviation Annual Income:", data['Annual Income (k$)'].std())
print("Standard Deviation Spending Score:", data['Spending Score (1-100)'].std())

# Range
print("Range Age:", data['Age'].max() - data['Age'].min())
print("Range Annual Income:", data['Annual Income (k$)'].max() - data['Annual Income (k$)'].min())
print("Range Spending Score:", data['Spending Score (1-100)'].max() - data['Spending Score (1-100)'].min())

# Interquartile Range (IQR)
print("IQR Age:", stats.iqr(data['Age']))
print("IQR Annual Income:", stats.iqr(data['Annual Income (k$)']))
print("IQR Spending Score:", stats.iqr(data['Spending Score (1-100)']))
```

5. Frequency Tables

5.1 For Gender:

```python
data['Genre'].value_counts()
```

```python
#5.2 For Age groups (Optional, for nice table)
bins = [18, 25, 35, 45, 55, 70]
labels = ['18-25', '26-35', '36-45', '46-55', '56-70']
data['Age_Group'] = pd.cut(data['Age'], bins=bins, labels=labels)

# Frequency Table
data['Age_Group'].value_counts()
```

6. Charts/Graphs
6.1 Histogram

```python
sns.histplot(data['Age'], kde=True, color='blue')
plt.title('Age Distribution')
plt.show()

sns.histplot(data['Annual Income (k$)'], kde=True, color='green')
plt.title('Annual Income Distribution')
plt.show()

sns.histplot(data['Spending Score (1-100)'], kde=True, color='orange')
plt.title('Spending Score Distribution')
plt.show()
```
