```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

loading dataset sheet of titanic

```python
df = pd.read_excel('/content/titanic3.xls')
```

Exploratory data analysis

```python
df.head() #view first row of data
```

```python
df.isnull().sum() #to see missing values
```

```python
#Univariate analysis means analyzing one column (variable) at a time
sns.histplot(df['age'], bins=30, kde=True)
plt.title("Age Distribution of Passengers")
plt.show()
```

```python
sns.countplot(x='survived', data=df)
plt.title("Survival Count (0 = No, 1 = Yes)")
plt.show()
```

```python
#Perform bipartite function: Compare two variables to find trends
sns.barplot(x='sex', y='survived', data=df)
plt.title("Survival Rate by Gender")
plt.show()
```

```python
sns.barplot(x='pclass', y='survived', data=df)
plt.title("Survival Rate by Passenger Class")
plt.show()
```
