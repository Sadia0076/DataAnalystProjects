```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Step 2: Load and Explore the Dataset

```python
df = pd.read_csv('/content/weatherHistory.csv')
```

```python
# Show the first few rows
df.head()
```

Step 3: Clean the Data

```python
# Check for missing values
df.isnull().sum()


```

```python
df['Precip Type'] = df['Precip Type'].fillna('unknown')  #fill the misssing value im precip type column as remainin column is filled without any missing values
```

Step 4: Plot daily temperature trend

```python
print(df.columns)

```

```python
# Create the plot
plt.figure(figsize=(20, 6))
plt.plot(df['Temperature (C)'], label='Daily Temperature', color='blue', linewidth=2)

# Add title and labels
plt.title('Daily Temperature Over Time', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Temperature (Â°C)', fontsize=14)

# Show the legend
plt.legend()

# Add gridlines with a subtle, lighter style
plt.grid(True, linestyle='--', linewidth=0.5, color='gray')

# Show the plot
plt.show()
```

Step 5: Add 7-Day Moving Average (Trend Line)

```python
df['7_day_avg'] = df['Temperature (C)'].rolling(window=7).mean()

plt.figure(figsize=(12,6))
plt.plot(df['Temperature (C)'], alpha=0.5, label='Daily Temperature')
plt.plot(df['7_day_avg'], color='orange', label='7-Day Moving Average')
plt.title("Smoothed Temperature Trend")
plt.xlabel("Date")
plt.ylabel("Temperature (Â°C)")
plt.legend()
plt.grid(True)
plt.show()
```

 Statistical Analysis using NumPy

```python
print("Mean Temp:", np.mean(df['Temperature (C)']))
print("Max Temp:", np.max(df['Temperature (C)']))
print("Min Temp:", np.min(df['Temperature (C)']))
print("Standard Deviation:", np.std(df['Temperature (C)']))
```
